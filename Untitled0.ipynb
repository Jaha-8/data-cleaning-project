```python
import pandas as pd

data = {
    "Customer_ID": [101,102,103,103],
    "Name": ["Asha","Raj","Aman","Aman"],
    "Age": [22,None,-5,-5],
    "Gender": ["F","M","M","M"],
    "City": ["Delhi","Mumbai","Delhi","Delhi"],
    "Salary": [25000,30000,500000,500000],
    "Join_Date": ["2022-05-01","2021/06/15","2020-07-20","2020-07-20"],
    "DOB": ["2002-01-10","2001-03-20","1999-08-15","1999-08-15"]
}

df = pd.DataFrame(data)
df
```

```python
print(df['Age'])
```

```python
# Convert Age to numeric safely
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Fill missing values
avg_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(avg_age)

df
```

```python
df = df[df['Age']>0]
df
```

```python
df = df.drop_duplicates()
df
```

```python
df['Join_Date'] = df['Join_Date'].astype(str)
df['Join_Date'] = pd.to_datetime(df['Join_Date'], format='mixed', errors='coerce')

df
```

```python
df.to_csv("cleaned_data.csv",index=False)
```
